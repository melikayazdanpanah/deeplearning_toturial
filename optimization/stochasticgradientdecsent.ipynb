What is Stochastic Gradient Descent (SGD)?
In traditional Gradient Descent (also called Batch Gradient Descent), the algorithm computes the gradient of the cost function using the entire dataset. This can be computationally expensive and slow, especially for large datasets.

Stochastic Gradient Descent (SGD) is a variation of gradient descent where the gradient is computed using only a single data point (or a small batch) at each step. This makes the updates much faster but introduces more noise in the learning process, leading to more frequent updates that can sometimes help the model escape local minima.

```python
import tensorflow as tf 

#data \\\ training
x_train = tf.constant([[1.0],[2.0],[3.0],[4.0]], dtype=tf.float32)
y_train = tf.constant([[2.0],[4.0],[6.0],[8.0]],dtype=tf.float32)

#our model
model = tf.keras.Sequential([tf.keras.layers.Dense(1, input_shape=(1,))])
#compile model
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate = 0.01),loss="mean_squared_error")
#train model
model.fit(x_train,y_train , epochs=100 , verbose=2)
```

Epoch 1/100: Indicates the first epoch out of 100. The initial loss is very high at 97.6981, meaning the model's predictions are initially far from accurate.
Epoch Progress: As training continues, the loss decreases significantly. For example:
Epoch 2: Loss drops to 67.8514.
Epoch 10: Loss reduces further to 3.8269.
Epoch 50: Loss reaches 0.1507.
Epoch 100: Final loss is 0.1117
